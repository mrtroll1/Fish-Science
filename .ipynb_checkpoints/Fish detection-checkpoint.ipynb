Now let's check (just in case) whether all the images are labeled and only keep the images with a single label (to focus on one object detection at a time). 

```python
def sanitize_dataframe(dataframe):

    no_class_counter = 0
    multiple_classes_counter = 0
    no_class_rows = []
    multiple_class_rows = []


    for index, row in dataframe.iterrows():
        category = row['objects']['category']

        if len(category) == 1:
            pass
        elif len(category) >= 2:
            no_class_counter += 1
            no_class_rows.append(index)
        else:
            multiple_classes_counter += 1
            multiple_class_rows.append(index)

    print(f'Number of rows with no classes: {no_class_counter}') 
    print(f'Number of rows with multiple classes: {multiple_classes_counter}')
    rows_to_drop = no_class_rows + multiple_class_rows

    dataframe.drop(rows_to_drop, axis=0, inplace=True)
    
    return dataframe

df = sanitize_dataframe(df)
```

Looks like that wasn't unnecessary. Finally, let's calculate the RGB-channels mean and standard deviation for future normalization.

```python
import numpy as np

means = []
stds = []

for index, row in df.iterrows():
    image = Image.open(io.BytesIO(row['image']['bytes']))
    image = image.convert('RGB')
    np_image = np.array(image) / 255.0 # Images were not scaled yet

    mean = np.mean(np_image, axis=(0, 1))
    std = np.std(np_image, axis=(0, 1))

    means.append(mean)
    stds.append(std)

overall_mean = np.round(np.mean(means, axis=0), 3)
overall_std = np.round(np.mean(stds, axis=0), 3)

print(f'Overall Mean (RGB): {overall_mean}')
print(f'Overall Std Dev (RGB): {overall_std}')
```

<h2>Building a fish classifier</h2>

Now that we have our data set up, it is time to learn! First, let's try solving the classification task.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleClassifierCNN(nn.Module):
    def __init__(self, num_classes=20):
        super(SimpleClassifierCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * 8 * 8, 512)  
        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 64 * 8 * 8)  
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import pandas as pd
import numpy as np
from PIL import Image
import io

class FishDataset(Dataset):
    def __init__(self, dataframe, task, transform=None):
        self.dataframe = dataframe
        self.transform = transform
        self.task = task

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = Image.open(io.BytesIO(row['image']['bytes']))
        label = row['objects']['category'] 
        box = row['objects']['bbox']
        

        if self.transform:
            image = self.transform(image)

        if isinstance(image, np.ndarray) and not image.flags['WRITEABLE']:
            image = np.array(image, copy=True)
        
        if self.task == 'classification':
            return image, label
        elif self.task == 'detection':
            return image, box[0]
        elif self.task == 'merged':
            return (image, (box[0], label))


transform = transforms.Compose([
    transforms.Resize((64, 64)),  
    transforms.ToTensor(),
    transforms.Normalize(mean=overall_mean, std=overall_std),
])

train_dataset = FishDataset(df, task='classification', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=20, shuffle=True)
```

```python
def evaluate_classifier(model, val_loader):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            labels = labels.squeeze()

            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Accuracy on validation set: {accuracy:.2f}%')

    return accuracy
```

```python
def train_classifier(model, train_loader, criterion, epochs=10, learning_rate=0.001):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            labels = labels.squeeze()
            
            optimizer.zero_grad()
            outputs = model(images)
            outputs = outputs.squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}], Loss: {round(total_loss/len(train_loader), 4)}")
        evaluate_classifier(model, val_loader)

    print("Training complete")
```

```python
criterion = nn.CrossEntropyLoss()
model = SimpleClassifierCNN(num_classes=20)
train_classifier(model, train_loader, criterion, epochs=5)
```

<h2>Building a fish detector</h2>

Now let's build a simple object detector. Not a single fish should hide! First, let's define a Generalized Intersection loss function. The model would train to maximize the overlap of predicted box and target box.

```python
def iou_loss(pred_boxes, target_boxes):
    pred_x2 = pred_boxes[:, 0] + pred_boxes[:, 2]
    pred_y2 = pred_boxes[:, 1] + pred_boxes[:, 3]
    target_x2 = target_boxes[:, 0] + target_boxes[:, 2]
    target_y2 = target_boxes[:, 1] + target_boxes[:, 3]

    inter_x1 = torch.max(pred_boxes[:, 0], target_boxes[:, 0])
    inter_y1 = torch.max(pred_boxes[:, 1], target_boxes[:, 1])
    inter_x2 = torch.min(pred_x2, target_x2)
    inter_y2 = torch.min(pred_y2, target_y2)

    inter_area = torch.clamp(inter_x2 - inter_x1, min=0) * torch.clamp(inter_y2 - inter_y1, min=0)
    pred_area = pred_boxes[:, 2] * pred_boxes[:, 3]
    target_area = target_boxes[:, 2] * target_boxes[:, 3]
    union_area = pred_area + target_area - inter_area

    iou_loss = 1 - (inter_area / union_area)

    return iou_loss.mean()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleDetectorCNN(nn.Module):
    def __init__(self, new_size):
        super(SimpleDetectorCNN, self).__init__()
        self.new_size = new_size
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * (self.new_size // 8) ** 2, 512)  
        self.fc2 = nn.Linear(512, 4)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 64 * (self.new_size // 8) ** 2)  
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = torch.sigmoid(x) * self.new_size  
        return x
```

```python
new_size = 64

transform = transforms.Compose([
    transforms.Resize((new_size, new_size)),  
    transforms.ToTensor(),
    transforms.Normalize(mean=overall_mean, std=overall_std),
])

train_dataset = FishDataset(df, task='detection', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=20, shuffle=False)

val_dataset = FishDataset(df_val, task='detection', transform=transform)
val_loader = DataLoader(val_dataset, batch_size=20, shuffle=False)
```

```python
def evaluate_detector(model, val_loader):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    total_loss = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            labels = labels.squeeze()
            labels = labels / 10

            outputs = model(images)
            outputs = outputs.squeeze()
            loss = iou_loss(outputs, labels)
            total_loss += loss.item()
            
    total_loss = total_loss / len(val_loader)      
    print(f'Loss on validation set: {total_loss:.4f}')

    return total_loss
```

```python
def train_detector(model, train_loader, epochs=5, learning_rate=0.001):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            labels = labels.squeeze()
            labels = labels / 10
            
            optimizer.zero_grad()
            outputs = model(images)
            outputs = outputs.squeeze()
            loss = iou_loss(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}], Loss: {round(total_loss/len(train_loader), 4)}")
        evaluate_detector(model, val_loader)

    print("Training complete")
```

```python
def unnormalize(tensor, mean, std):
    mean = torch.as_tensor(mean, dtype=tensor.dtype, device=tensor.device)
    std = torch.as_tensor(std, dtype=tensor.dtype, device=tensor.device)
    
    mean = mean[:, None, None]
    std = std[:, None, None]

    return tensor * std + mean
```

<h2>Combining two models</h2>

Let's try merging two models into one. The 'path' for box and label outputs will be separate after all the convolutions and pooling. That way, we avoid the significant increase in the number of parameters but ensure the learning of both components.

```python
class SimpleMergedCNN(nn.Module):
    def __init__(self, num_channels=16, kernel_size=3, stride=1, new_size=64, num_classes=20, lin_layer_size=512):
        super(SimpleMergedCNN, self).__init__()
        self.new_size = new_size
        self.num_channels = num_channels
        self.conv1 = nn.Conv2d(3, num_channels, kernel_size=kernel_size, stride=stride, padding=kernel_size//2)
        self.conv2 = nn.Conv2d(num_channels, num_channels * 2, kernel_size=kernel_size, stride=stride, padding=kernel_size//2)
        self.conv3 = nn.Conv2d(num_channels * 2, num_channels * 4, kernel_size=kernel_size, stride=stride, padding=kernel_size//2)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1_0 = nn.Linear(num_channels * 4 * (self.new_size // 8) ** 2, lin_layer_size) 
        self.fc1_1 = nn.Linear(num_channels * 4 * (self.new_size // 8) ** 2, lin_layer_size) 
        self.fc2_0 = nn.Linear(lin_layer_size, 4)
        self.fc2_1 = nn.Linear(lin_layer_size, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 4 * self.num_channels * (self.new_size // 8) ** 2)
        box = F.relu(self.fc1_0(x))
        label = F.relu(self.fc1_1(x))
        box = self.fc2_0(box)
        box = torch.sigmoid(box) * self.new_size 
        label = self.fc2_1(label)

        return box, label
```

```python
train_dataset = FishDataset(df, task='merged', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=20, shuffle=True)

val_dataset = FishDataset(df_val, task='merged', transform=transform)
val_loader = DataLoader(train_dataset, batch_size=20, shuffle=True)
```

```python
def evaluate_merged_model(model, val_loader, box_criterion):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    total_loss = 0
    total_count = 0
    correct = 0

    with torch.no_grad():
        for images, (bbox_labels, class_labels) in train_loader:
            images, bbox_labels, class_labels = images.to(device), bbox_labels.to(device), class_labels.to(device)
            bbox_preds, class_preds = model(images)
            bbox_labels = bbox_labels / 10
            
            bbox_preds = bbox_preds.squeeze()
            class_preds = class_preds.squeeze()
            class_labels = class_labels.squeeze()
            
            bbox_loss = box_criterion(bbox_preds, bbox_labels)
            
            _, class_preds = torch.max(class_preds.data, 1)
            
            total_count += class_labels.size(0)
            correct += (class_preds == class_labels).sum().item()
            total_loss += bbox_loss.item()
            
    total_loss = total_loss / len(val_loader)
    accuracy = 100 * correct / total_count
    print(f'Classification accuracy on val set: {accuracy:.4f}, Detection loss on val set: {total_loss:.4f}')

    return total_loss
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader


def train_merged_model(model, train_loader, val_loader, criterion, box_criterion, epochs=5, learning_rate=0.001):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate) 

    for epoch in range(epochs):
        model.train()
        total_bbox_loss = 0
        total_class_loss = 0

        for images, (bbox_labels, class_labels) in train_loader:
            images, bbox_labels, class_labels = images.to(device), bbox_labels.to(device), class_labels.to(device)

            optimizer.zero_grad()
            bbox_preds, class_preds = model(images)
            bbox_labels = bbox_labels / 10
            
            bbox_preds = bbox_preds.squeeze()
            class_preds = class_preds.squeeze()
            class_labels = class_labels.squeeze()
            
            bbox_loss = box_criterion(bbox_preds, bbox_labels)
            class_loss = class_criterion(class_preds, class_labels)
            loss = 0.9 * bbox_loss + 0.1 * class_loss # Balancing the task weigths
            loss.backward()
            optimizer.step()

            total_bbox_loss += bbox_loss.item()
            total_class_loss += class_loss.item()

        avg_bbox_loss = total_bbox_loss / len(train_loader)
        avg_class_loss = total_class_loss / len(train_loader)
        print(f"Epoch [{epoch+1}/{epochs}], Classification loss: {avg_class_loss:.4f}, Detection Loss: {avg_bbox_loss:.4f}")
        print('\n')
        evaluate_merged_model(model, val_loader, box_criterion)

    print("Training complete")
```

```python
class_criterion = nn.CrossEntropyLoss()
box_criterion = iou_loss
model = SimpleMergedCNN(lin_layer_size=512)
train_merged_model(model, train_loader, val_loader, class_criterion, box_criterion)
```

<h2>Reducing the size of the model</h2>

Let's see how sensitive our model is to killing small weigths.

```python
def prune_weights(layer, threshold):
    with torch.no_grad():
        weight_mask = torch.abs(layer.weight) > threshold
        layer.weight.data.mul_(weight_mask)
        if layer.bias is not None:
            bias_mask = torch.abs(layer.bias) > threshold
            layer.bias.data.mul_(bias_mask)
```

```python
import copy

model_to_prune = copy.deepcopy(model)

threshold = 0.1

prune_weights(model_to_prune.conv1, threshold)
prune_weights(model_to_prune.conv2, threshold)
prune_weights(model_to_prune.conv3, threshold)
prune_weights(model_to_prune.fc1_0, threshold)
prune_weights(model_to_prune.fc1_1, threshold)
prune_weights(model_to_prune.fc2_0, threshold)
prune_weights(model_to_prune.fc2_1, threshold)
```

```python
evaluate_merged_model(model_to_prune, val_loader, box_criterion);
```

The model is basically useless now. Let's continue exploring

```python
def calculate_percentage_of_small_weights(model, threshold):
    total_weights = 0
    small_weights = 0

    for module in model.modules():
        if isinstance(module, nn.Conv2d) or isinstance(module, nn.Linear):
            weights = module.weight.data

            total_weights += weights.numel()
            small_weights += (weights.abs() <= threshold).sum().item()

    percentage = 100 * small_weights / total_weights
    print(f"Out of {total_weights} weigths {percentage:.2f}% are smaller than {threshold_value}")
    
    return total_weights, percentage

threshold = 0.1 
calculate_percentage_of_small_weights(model, threshold);

threshold = 0.05
calculate_percentage_of_small_weights(model, threshold);

threshold = 0.01
calculate_percentage_of_small_weights(model, threshold);
```

Ok. Let's try killing with threshold of 0.01.

```python
import copy

model_to_prune = copy.deepcopy(model)

threshold = 0.01

prune_weights(model_to_prune.conv1, threshold)
prune_weights(model_to_prune.conv2, threshold)
prune_weights(model_to_prune.conv3, threshold)
prune_weights(model_to_prune.fc1_0, threshold)
prune_weights(model_to_prune.fc1_1, threshold)
prune_weights(model_to_prune.fc2_0, threshold)
prune_weights(model_to_prune.fc2_1, threshold)
```

```python
evaluate_merged_model(model_to_prune, val_loader, box_criterion);
```

Ok, nice! Looks like 42% of the weights could be harmlessly killed. Let's try building a model with simpler arhcitecture.

First instinct tells to keep the three convolutions and simply reduce the number of neurons in linear layers. The linear layers are responsible for the vast majority of weights anyway. Let's try this.
